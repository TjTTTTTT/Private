```python
import matplotlib.pyplot as plt

# 数据
sizes = [65, 35]
labels = ['Others', 'JIA']
colors = ['#B0E0E6', '#FFDAB9']  # 柔和的颜色：浅蓝和浅橙
explode = (0, 0.1)  # 突出显示第二个部分

# 创建饼图
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes,
                                  explode=explode,
                                  labels=labels,
                                  colors=colors,
                                  autopct='%1.1f%%',
                                  shadow=True,
                                  startangle=90,
                                  pctdistance=0.7,
                                  labeldistance=1.1)  # 将标签位置稍微放远一点，以避免重叠

# 设置等轴比以确保饼图是圆的
ax.axis('equal')

# 显示图表
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子以确保重现性
np.random.seed(0)

# 生成模拟数据
weeks = [f"Week{i}" for i in range(15, 25)]
coding = np.zeros(10)  # 初始化coding时间为0
learning = np.random.randint(2, 5, size=10)  # learning 替代 analysis
writing = np.zeros(10)  # 初始化写作时间为0
meeting = np.random.uniform(2.8, 3.2, size=10)  # 设置会议时间近似为3小时

# 特定周的coding时间调整
coding[:5] = np.random.uniform(2.5, 3.5, size=5)  # 第15到19周coding时间
coding[6:9] = np.random.uniform(3, 3.5, size=3)  # 第21到23周coding时间
# 第20周和第24周coding时间已初始化为0

# 特定周的写作和开会时间调整
writing[5] = 1.5  # 第20周写作时间1.5小时
writing[7:10] = np.random.uniform(2.5, 3, size=3)  # 第22到24周写作时间
meeting[7:10] = np.random.uniform(1.5, 2, size=3)  # 第22到24周开会时间调整为1.5到2小时

data = {
    'Week': weeks,
    'Coding': coding,
    'Learning': learning,
    'Writing': writing,
    'Meeting': meeting
}

df = pd.DataFrame(data)

# 柔和颜色定义
colors = ['#6897bb', '#ffc266', '#cc99ff', '#ffcc99']  # 更柔和的颜色设置

# 绘图
fig, ax = plt.subplots(figsize=(10, 6))

# 累加前面的活动时间，以便堆叠
bottom = np.zeros(len(df))

activities = ['Coding', 'Learning', 'Writing', 'Meeting']
for activity, color in zip(activities, colors):
    ax.bar(df['Week'], df[activity], bottom=bottom, color=color, label=activity)
    bottom += df[activity].values

# 添加标题和标签
# ax.set_title("Jia's Activity")
ax.set_xlabel("Weeks")
ax.set_ylabel("Hours")
ax.legend(title="Activity")

# 展示图表
plt.show()
```
